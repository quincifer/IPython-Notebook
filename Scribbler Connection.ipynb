## Connecting to Scribbler Robot using an IPython Notebook

Scribbler robots use the Myro Library, which comes packaged in the Calico IDE. There are several advantages, however, to using IPython notebooks over Calico in a teaching milieu. Short of contacting the IPython project and encouraging them to add the Myro library to their package, the best way to make Scribbler robots compatible on notebooks is by manually installing myro and its dependencies.

Follow these steps make IPython compatible with the myro library (Mac OS X):<br>
1) Download the latest version of myro from http://myro.roboteducation.org/download/<br>
2) Download PySerial from https://pypi.python.org/pypi/pyserial and download the -tar.gz file<br>
3) Open the Downloads folder and unzip both files<br>
4) Open /Downloads/myro-2.9.1/myro/_ _init_ _.py in text edit, or windows compatable text editor<br>
6) Scroll to line 1589 (or command+L -> 1589) and remove the command to import Epuck. Reference picture<br>
7) cd into /Downloads/pyserial-2.7 (or other version) and run setup.py with the command (python setup.py install)<br>
8) cd into /Downloads/myro-2.9.1 through terminal and run setup.py with the command (python setup.py install)<br>
9) Conda install pillow

```python
#Open saved images png files in a separate viewing window
#From PIL import Image
im = Image.open("/Users/quincy/picture.png")
im.show()
```

```python
#reads in a picture
p = makePicture("/Users/quincy/picture.png")
#p.image creates a PIL image taken from a myro fluke2 picture
image = p.image
#creates another picture using a PIL rotate command(degrees)
out = image.rotate(45)
#displays PIL image in separate viewing window
out.show()
```

```python
#import numpy
#creates an array of PIL image
arr = numpy.array(image)
#prints size of array
arr.shape
```

```python
#manipulates second to top, three rightmost pixels to be pink
arr[0:80, 0:128, 0] = (255)
arr[0:80, 0:128, 1] = (105)
arr[0:80, 0:128, 2] = (180)
#creates a PIL image from numpy array
img = Image.fromarray(arr)
img.save("output1.png")
```

```python
#transpose array
arrt = arr.transpose()
#show new dimensions
arrt.shape
```

```python
pil_im = Image.open("output1.png")
imshow(numpy.array(pil_im))
```

```python
#load saved picture into an open cv array which reads bgr, not rgb
#show using matplotlib command
#im_cv is an array
im_cv = cv2.imread("output1.png")[:,:,:]
imshow(im_cv)
```

```python
#Convert to RGB
im_cv = im_cv[:, :, ::-1]
imshow(im_cv)
```
