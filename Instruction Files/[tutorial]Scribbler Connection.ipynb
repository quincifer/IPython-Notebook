```python
#display inline scribbler image as PIL image without saving to hard drive using matplotlib inline
backward(1, 1)
forward(1, 1)
pic_inline = takePicture()
PIL_inline = pic_inline.image
array_inline = numpy.array(PIL_inline)
imshow(array_inline)
```

```python
#Take three more images oriented 60 degrees apart and convert them to PIL images and then to numpy arrays
turnBy(40, "deg")
pic1_inline = takePicture()
PIL1_inline = pic1_inline.image
array1_inline = numpy.array(PIL1_inline)
turnBy(40, "deg")
pic2_inline = takePicture()
PIL2_inline = pic2_inline.image
array2_inline = numpy.array(PIL2_inline)
turnBy(40, "deg")
pic3_inline = takePicture()
PIL3_inline = pic3_inline.image
array3_inline = numpy.array(PIL3_inline)
```

```python
#concatenate the images along the first axis (horizontally) and display
total = numpy.concatenate((array_inline, array1_inline, array2_inline, array3_inline), axis=1)
imshow(total)
```
